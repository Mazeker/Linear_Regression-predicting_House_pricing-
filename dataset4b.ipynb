```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("new2.csv")
data.info()
```

```python
data["Zipcode"]=data["Zipcode"].astype(object)
```

```python
data.dtypes
```

```python
data['No of Times Visited'].unique()
```

```python
mapping={'None':'0',
        'Once':'1',
        'Twice':'2',
        'Thrice':'3',
        'Four':'4'}
data['No of Times Visited']=data['No of Times Visited'].map(mapping)
```

```python
data['No of Times Visited'].unique()
```

```python
data.head()
```

```python
data['Ever Renovate']=np.where(data['Renovated Year']==0,'No','Yes')
```

```python
data.head()
```

```python
data['Purchase Year']=pd.DatetimeIndex(data['Date House was Sold']).year
data.head()
```

```python
data['Years Since Renovation']=np.where(data['Ever Renovate']=='Yes',abs(data['Purchase Year']-data['Renovated Year']),0)
```

```python
data.head()
```

```python
data.drop(columns=['Renovated Year','Date House was Sold','Purchase Year'],inplace=True)
```

```python
data.head()
```

```python
data.to_csv("New3",index=False)
```
